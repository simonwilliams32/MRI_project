```python
import numpy as np
from PIL import Image
from tensorflow.keras.applications import efficientnet

IMG_SIZE = (224, 224)

def preprocess_image(image_path):
    """
    this function is used for the processing of the image provided by the user
    """

    img = Image.open(image_path).convert("RGB")

    # resize
    img = img.resize(IMG_SIZE, Image.LANCZOS)

    # convert to np.array
    arr = np.array(img).astype(np.float32)

    # scale using the efficentnet
    arr = efficientnet.preprocess_input(arr)

    # Add batch size (1)
    arr = np.expand_dims(arr, axis=0)

    return arr
```

```python
import tensorflow as tf

model = tf.keras.models.load_model("brain_tumor_model_4class_transferlearning.keras")
class_names = ["glioma", "meningioma", "notumor", "pituitary"]

x = preprocess_image("/home/onuro/code/simonwilliams32/MRI_project/test_images/m.png")
preds = model.predict(x, verbose=0)[0]

predicted_idx = int(np.argmax(preds))
print("Predicted class:", class_names[predicted_idx])
print("Confidence:", preds[predicted_idx])
print("All probabilities:", dict(zip(class_names, preds)))
```
